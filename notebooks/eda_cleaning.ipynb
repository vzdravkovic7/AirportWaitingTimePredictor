```python
# Učitavanje biblioteka
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Učitavanje podataka (spojeni CSV fajlovi)
import glob

files = glob.glob("../data/*.csv")  # putanja do data foldera
dfs = [pd.read_csv(f) for f in files]
df = pd.concat(dfs, ignore_index=True)

print("Shape:", df.shape)
df.head()
```

```python
# Osnovne informacije o podacima
df.info()
df.describe(include="all").T
```

```python
# Provera nedostajućih vrednosti i duplikata
print("Missing values:\n", df.isnull().sum())
print("Duplicates:", df.duplicated().sum())
```

```python
# Konverzija datuma
df['FlightDate'] = pd.to_datetime(df['FlightDate'])

# Dodavanje novih kolona
df['day_of_week'] = df['FlightDate'].dt.dayofweek   # 0=Monday
df['month'] = df['FlightDate'].dt.month
df['year'] = df['FlightDate'].dt.year

# Sezona
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

df['season'] = df['month'].apply(get_season)
```

```python
# Vizuelizacija distribucija
plt.figure(figsize=(8,5))
sns.histplot(df['AverageWait'], bins=50, kde=True)
plt.title("Distribucija prosečnog čekanja")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='AirportCode', y='AverageWait', data=df)
plt.title("Prosečno čekanje po aerodromu")
plt.show()
```

```python
# Čišćenje podataka
# Opcija 1: Izbaciti redove gde nema putnika
# df = df[df['TotalPassengerCount'] > 0]

# Opcija 2: Izbaciti nerelevantne redove sa 0 AverageWait
df = df[df['AverageWait'] > 0]

# Izbaciti duplikate
df = df.drop_duplicates()

print("Shape nakon čišćenja:", df.shape)
```

```python
# Snimanje očišćenog dataset-a
df.to_csv("../data/cleaned_airport_waits.csv", index=False)
print("Očišćen dataset snimljen u data/cleaned_airport_waits.csv")
```
